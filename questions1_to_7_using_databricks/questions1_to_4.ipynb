```python
import requests
import json
import pandas as pd
from pyspark.sql.functions import col, lit, current_timestamp, expr

def get_all_coins_list(api_key):
    url = 'https://api.coingecko.com/api/v3/coins/list'
    params = {  
            'vs_currencies': 'USD'
    }

    # Replace 'YOUR_API_KEY' with your actual API key
    headers = { 'x-cg-demo-api-key': api_key }
    response = requests.get(url, params = params)

    # Convert JSON string to Python dictionary
    json_data = json.dumps(response.json(), indent=4)
    dataprep = json.loads(json_data)

    # Convert dictionary to DataFrame
    #dfspark = spark.createDataFrame(pd.DataFrame(dataprep))
    dfspark = spark.createDataFrame(dataprep)
    return dfspark
```

```python
# 3.-Retrieve a list of all vavailable coins in the API
all_coins=get_all_coins_list('CG-HwZEyrLTyMJR2hhdmCvxFurs')
all_coins.display()
```
