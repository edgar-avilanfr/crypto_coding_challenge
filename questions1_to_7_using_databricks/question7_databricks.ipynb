```python
from pyspark.sql.functions import col, lit, current_timestamp, expr, avg
from pyspark.sql import Window

bit_prices=spark.sql("""
select *
from vistra_tst_worspace3.test_schema2.bitcoin_price_historical
          """).withColumn('ID', lit('Bitcoin'))

window_spec = Window.partitionBy('ID').orderBy('date').rowsBetween(-5, Window.currentRow)
bit_prices =bit_prices.withColumn('moving_avg', avg(col('price')).over(window_spec))
```
